```python
train.head()
```

# Bar Plot

```python
sns.barplot(x='Pclass', y='Age', data=train)
```

```python
sns.barplot(x='Pclass', y='Fare', data=train)
```

```python
sns.barplot(x='Pclass', y='Fare', data=train, hue='Sex')
```

# BoxPlot

```python
sns.boxplot(x='Sex',y='Age', data=train, hue='Survived')
```

# Dist Plot

```python
sns.distplot(train[train['Survived']==0]['Age'],hist=False)
sns.distplot(train[train['Survived']==1]['Age'],hist=False)
```

# Heat Map

```python
pd.crosstab(train['Pclass'],train['Survived'])
```

```python
sns.heatmap(pd.crosstab(train['Pclass'],train['Survived']))
```

# Cluster Map

```python
pd.crosstab(train['SibSp'],train['Survived'])
```

```python
sns.clustermap(pd.crosstab(train['SibSp'],train['Survived']))
```

# pair Plot

```python
sns.pairplot(iris)
```
